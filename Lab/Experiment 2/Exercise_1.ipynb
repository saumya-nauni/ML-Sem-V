# **EXPERIMENT 2**

**EXERCISE 1: INTRODUCTION AND BACKGROUND**

**A. Importing dataset as dataframe in python.**




```
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```
filename = "https://s3-api.us-geo.objectstorage.softlayer.net/cf-courses-data/CognitiveClass/DA0101EN/auto.csv"
```

**B. Add column headers to dataset.**





```
headers = ["symboling","normalized-losses","make","fuel-type","aspiration", "num-of-doors","body-style",
"drive-wheels","engine-location","wheel-base", "length","width","height","curb-weight","engine-type",
"num-of-cylinders", "engine-size","fuel-system","bore","stroke","compression-ratio","horsepower",
"peak-rpm","city-mpg","highway-mpg","price"]
```

**C. Read first five lines and display.**

```
data=pd.read_csv(filename, names=headers)
```

```
data.head()
```

**D. Are there missing data or illegal characters in the dataset?**

As there are "?" present in the above table we can assume that there are missing values present in the dataset

**E. Replace "?" with nan in the dataset.**

```
data=data.replace("?",np.nan)
```

**F. Count missing values in each column and print it with column name.**

```
data.isnull().sum()
```

From the above output we can infer that the missing values are as follows:


> normalized-losses:    41

> num-of-doors:    2

> bore :   4

> stroke: 4

> horsepower: 2

> peak-rpm: 2

> price: 4



**G. Delete price rows that have missing data.**

```
data.dropna(subset = ['price'],inplace = True)
data
```


**H. Replace the missing values in the columns of normalized losses, bore, stroke, horsepower and peak-rpm with the mean.**

```
data['normalized-losses'].replace(np.nan,(data['normalized-losses'].astype('float').mean(axis=0)),inplace = True)
data['bore'].replace(np.nan,(data['bore'].astype('float').mean(axis=0)),inplace = True)
data['stroke'].replace(np.nan,(data['stroke'].astype('float').mean(axis=0)),inplace = True)
data['horsepower'].replace(np.nan,(data['horsepower'].astype('float').mean(axis=0)),inplace = True)
data['peak-rpm'].replace(np.nan,(data['peak-rpm'].astype('float').mean(axis=0)),inplace = True)
```

**I. Num-of-doors: replace missing values with most frequent value in the column.**

```
data['num-of-doors'].replace(np.nan,data['num-of-doors'].value_counts().idxmax(),inplace = True )
data
```

**J. Reset the index of the dataframe**

```
data.reset_index()
```

**K. Check datatype of columns and convert numeric/quantitative variables to float or int.**

```
data.dtypes
```

```
for i in data._get_numeric_data().columns:
    data[i] = data[i].astype('float')
data.info()
```

**L. Transform city-mpg and highway-mpg into litres/100km using conversion formula.**

```
data['city-L/100km'] = 235/data['city-mpg']
data['highway-L/100km'] = 235/data['highway-mpg']
data
```

**M. Normalize columns length, width, height so that their values range from 0 to 1.**

```
data['length'] = data['length']/data['length'].max()
data['width']=data['width']/data['width'].max()
data['height']=data['height']/data['height'].max()
data
```

**N. Plot the histogram of horsepower to see its distribution.**

```
data['horsepower'].plot(kind='hist')
```

**O. Create 3 equal sized bins 'low', 'medium' and 'high' and organize values from column horsepower into new column 'horsepower-binned'.**

```
data['horsepower-binned']=pd.cut(data['horsepower'], bins=3, labels=['low', 'medium', 'high'])
data
```

**P. Plot distribution of 'horsepower-binned'.**

```
plt.hist(data['horsepower-binned'])
```

**Q. Convert 'fuel-type' into one-hot-encoded variables. Repeat same for 'aspiration' and then drop the columns of 'fuel-type' and 'aspiration'.**

```
one_hot1 = pd.get_dummies(data['fuel-type'])
one_hot2 = pd.get_dummies(data['aspiration'])
data = data.join(one_hot1)
data = data.join(one_hot2)
data.drop("fuel-type", axis = 1, inplace=True)
data.drop("aspiration", axis = 1, inplace=True)
data
```
